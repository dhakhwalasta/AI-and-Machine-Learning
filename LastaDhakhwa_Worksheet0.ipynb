Task - 1:

Create a Python program that converts between different units of measurement.

• The program should:
1. Prompt the user to choose the type of conversion (e.g., length, weight, volume).
2. Ask the user to input the value to be converted.
3. Perform the conversion and display the result.
4. Handle potential errors, such as invalid input or unsupported conversion types.

• Requirements:

1. Functions: Define at least one function to perform the conversion.
2. Error Handling: Use try-except blocks to handle invalid input (e.g., non-numeric values).
3. User Input: Prompt the user to select the conversion type and input the value.
4. Docstrings: Include a docstring in your function to describe its purpose, parameters, and
return value.

• Conversion Options:

1. Length:

– Convert meters (m) to feet (ft).

– Convert feet (ft) to meters (m).

2. Weight:

– Convert kilograms (kg) to pounds (lbs).

– Convert pounds (lbs) to kilograms (kg).

3. Volume:

– Convert liters (L) to gallons (gal).

– Convert gallons (gal) to liters (L).


```python
def convert_units(choice, value):
    """
    Converts a given value based on the selected conversion type.

    Parameters:
    choice (str): The type of conversion to perform.
                           Options:
                           '1' - Meters to Feet
                           '2' - Feet to Meters
                           '3' - Kilograms to Pounds
                           '4' - Pounds to Kilograms
                           '5' - Liters to Gallons
                           '6' - Gallons to Liters
    value (float): The numeric value to convert.

    Returns:
    float: The converted value.
    """

    if choice == '1':
        return value * 3.28084  # meters to feet
    elif choice == '2':
        return value / 3.28084  # feet to meters
    elif choice == '3':
        return value * 2.20462  # kg to pounds
    elif choice == '4':
        return value / 2.20462  # pounds to kg
    elif choice == '5':
        return value * 0.264172  # liters to gallons
    elif choice == '6':
        return value / 0.264172  # gallons to liters
    else:
        return None


def main():
    print("Select Conversion Type:")
    print("1. Meters (m) to Feet (ft)")
    print("2. Feet (ft) to Meters (m)")
    print("3. Kilograms (kg) to Pounds (lbs)")
    print("4. Pounds (lbs) to Kilograms (kg)")
    print("5. Liters (L) to Gallons (gal)")
    print("6. Gallons (gal) to Liters (L)")

    choice = input("Enter your choice (1-6): ")

    try:
        value = float(input("Enter the value to convert: "))

        result = convert_units(choice, value)

        if result is None:
            print("Error: Unsupported conversion type selected.")
        else:
            print(f"Converted value: {result}")

    except ValueError:
        print("Error: Please enter a valid numeric value.")

main()
```

Create a Python program that performs various mathematical operations on a list of numbers.

• The Program should:

1. Prompt the user to choose an operation (e.g., find the sum, average, maximum, or minimum
of the numbers).
2. Ask the user to input a list of numbers (separated by spaces).

• Requirements:

1. Functions: Define at least one function for each operation (sum, average, maximum, minimum).
2. Error Handling: Use try-except blocks to handle invalid input (e.g., non-numeric values or
empty lists).
3. User Input: Prompt the user to select the operation and input the list of numbers.
4. Docstrings: Include a docstring in each function to describe its purpose, parameters, and

```python
def calculate_sum(numbers):
    """
    Calculates the sum of a list of numbers.

    Parameters:
    numbers (list of float): A list containing numeric values.

    Returns:
    float: The total sum of the numbers.
    """
    return sum(numbers)


def calculate_average(numbers):
    """
    Calculates the average of a list of numbers.

    Parameters:
    numbers (list of float): A list containing numeric values.

    Returns:
    float: The average of the numbers.
    """
    return sum(numbers) / len(numbers)


def find_maximum(numbers):
    """
    Finds the maximum value in a list of numbers.

    Parameters:
    numbers (list of float): A list containing numeric values.

    Returns:
    float: The largest number in the list.
    """
    return max(numbers)


def find_minimum(numbers):
    """
    Finds the minimum value in a list of numbers.

    Parameters:
    numbers (list of float): A list containing numeric values.

    Returns:
    float: The smallest number in the list.
    """
    return min(numbers)


def main():
    print("Choose an operation:")
    print("1. Find Sum")
    print("2. Find Average")
    print("3. Find Maximum")
    print("4. Find Minimum")

    choice = input("Enter your choice (1-4): ")

    try:
        user_input = input("Enter numbers separated by spaces: ")
        number_list = [float(num) for num in user_input.split()]

        if len(number_list) == 0:
            print("Error: The list cannot be empty.")
            return

        if choice == '1':
            result = calculate_sum(number_list)
            print(f"Sum: {result}")

        elif choice == '2':
            result = calculate_average(number_list)
            print(f"Average: {result}")

        elif choice == '3':
            result = find_maximum(number_list)
            print(f"Maximum: {result}")

        elif choice == '4':
            result = find_minimum(number_list)
            print(f"Minimum: {result}")

        else:
            print("Error: Invalid operation selected.")

    except ValueError:
        print("Error: Please enter only numeric values separated by spaces.")
    main()
```

4.3 Exercise on Nested List:

1. Flatten a Nested List:

Write a Python function that takes a nested list and flattens it into a single list, where all the elements
are in a single dimension.

• Requirements:

– Define a function flatten(lst) that takes a nested list lst and returns a flattened version
of the list.

– Example: For the input [[1, 2], [3, 4], [5]], the output should be [1, 2, 3, 4, 5].

```python
def flatten(lst):
    """
    Flattens a nested list into a single list (one-dimensional).

    Parameters:
    lst (list): A nested list (list of lists).

    Returns:
    list: A flattened list containing all elements in a single dimension.
    """
    flat_list = []
    for sublist in lst:
        for item in sublist:
            flat_list.append(item)
    return flat_list

nested_list = [[1, 2], [3, 4], [5]]
result = flatten(nested_list)

print("Original Nested List:", nested_list)
print("Flattened List:", result)
```

Write a Python function that extracts a specific element from a nested list given its indices.

```python
def access_nested_element(lst, indices):
    """
    Accesses an element in a nested list using a list of indices.

    Parameters:
    lst (list): The nested list.
    indices (list of int): A list of indices specifying the path to the element.

    Returns:
    any: The element located at the given indices.
    """
    element = lst
    try:
        for index in indices:
            element = element[index]
        return element
    except (IndexError, TypeError):
        raise ValueError("Invalid indices for the nested list.")

nested_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
indices = [1, 2]
result = access_nested_element(nested_list, indices)

print("Nested List:", nested_list)
print(f"Element at indices {indices}:", result)
```

3. Sum of All Elements in a Nested List:

Write a Python function that calculates the sum of all the numbers in a nested list (regardless of depth).

• Requirements:

– Define a function sum nested(lst) that takes a nested list lst and returns the sum of all
the elements.

– Example: For the input [[1, 2], [3, [4, 5]], 6], the output should be 21

```python
def sum_nested(lst):
    """
    Calculates the sum of all numbers in a nested list, regardless of depth.

    Parameters:
    lst (list): A nested list containing numbers or other lists.

    Returns:
    int/float: The sum of all numbers in the nested list.
    """
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)  #recursive
        else:
            total += item
    return total

nested_list = [[1, 2], [3, [4, 5]], 6]
result = sum_nested(nested_list)

print("Nested List:", nested_list)
print("Sum of All Elements:", result)
```

```python
def remove_element(lst, elem):
    """
    Removes all occurrences of a specific element from a nested list.

    Parameters:
    lst (list): A nested list (list of lists).
    elem (any): The element to remove.

    Returns:
    list: A new nested list with the element removed from all sublists.
    """
    modified_list = []
    for sublist in lst:
        if isinstance(sublist, list):
            filtered_sublist = [x for x in sublist if x != elem]
            modified_list.append(filtered_sublist)
        else:
            # If the top-level element itself matches elem, skip it
            if sublist != elem:
                modified_list.append(sublist)
    return modified_list

nested_list = [[1, 2], [3, 2], [4, 5]]
elem_to_remove = 2
result = remove_element(nested_list, elem_to_remove)

print("Original Nested List:", nested_list)
print(f"Nested List after removing {elem_to_remove}:", result)
```

5. Find the Maximum Element in a Nested List:

Write a Python function that finds the maximum element in a nested list (regardless of depth).

• Requirements:

– Define a function find max(lst) that takes a nested list lst and returns the maximum
element.

– Example: For the input [[1, 2], [3, [4, 5]], 6], the output should be 6.

```python
def find_max(lst):
    """
    Finds the maximum element in a nested list, regardless of depth.

    Parameters:
    lst (list): A nested list containing numbers or other lists.

    Returns:
    int/float: The maximum number in the nested list.
    """
    max_value = float('-inf')
    for item in lst:
        if isinstance(item, list):
            max_value = max(max_value, find_max(item))  # recursive
        else:
            max_value = max(max_value, item)
    return max_value

nested_list = [[1, 2], [3, [4, 5]], 6]
result = find_max(nested_list)

print("Nested List:", nested_list)
print("Maximum Element:", result)
```

6. Count Occurrences of an Element in a Nested List:

Write a Python function that counts how many times a specific element appears in a nested list.

• Requirements:

– Define a function count occurrences(lst, elem) that counts the occurrences of elem in
the nested list lst.

– Example: For the input lst = [[1, 2], [2, 3], [2, 4]] and elem = 2, the output should
be 3.

```python
def count_occurrences(lst, elem):
    """
    Counts how many times a specific element appears in a nested list.

    Parameters:
    lst (list): A nested list containing numbers or other lists.
    elem (any): The element to count.

    Returns:
    int: The total number of occurrences of the element.
    """
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)  # recursive
        else:
            if item == elem:
                count += 1
    return count

nested_list = [[1, 2], [3, 2], [2, 5]]
elem_to_count = 2
result = count_occurrences(nested_list, elem_to_count)

print("Nested List:", nested_list)
print(f"Occurrences of {elem_to_count}:", result)
```

7. Flatten a List of Lists of Lists:

Write a Python function that flattens a list of lists of lists into a single list, regardless of the depth.

• Requirements:

– Define a function deep flatten(lst) that takes a deeply nested list lst and returns a single
flattened list.

– Example: For the input [[[1, 2], [3, 4]], [[5, 6], [7, 8]]], the output should be
[1, 2, 3, 4, 5, 6, 7, 8].


```python
def deep_flatten(lst):
    """
    Flattens a deeply nested list into a single list, regardless of depth.

    Parameters:
    lst (list): A nested list of any depth.

    Returns:
    list: A flattened list containing all elements in a single dimension.
    """
    flat_list = []
    for item in lst:
        if isinstance(item, list):
            flat_list.extend(deep_flatten(item))  # recursive flatten
        else:
            flat_list.append(item)
    return flat_list

nested_list = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
result = deep_flatten(nested_list)

print("Original Nested List:", nested_list)
print("Deep Flattened List:", result)
```

8. Nested List Average:

Write a Python function that calculates the average of all elements in a nested list.

• Requirements:

– Define a function average nested(lst) that takes a nested list lst and returns the average
of all the elements.

– Example: For the input [[1, 2], [3, 4], [5, 6]], the output should be 3.5.


```python
def average_nested(lst):
    """
    Calculates the average of all elements in a nested list, regardless of depth.

    Parameters:
    lst (list): A nested list containing numbers or other lists.

    Returns:
    float: The average of all numeric elements.
    """
    def flatten_and_sum_count(sublist):
        total = 0
        count = 0
        for item in sublist:
            if isinstance(item, list):
                t, c = flatten_and_sum_count(item)  # recursive
                total += t
                count += c
            else:
                total += item
                count += 1
        return total, count

    total_sum, total_count = flatten_and_sum_count(lst)
    if total_count == 0:
        return 0  # avoid division by zero
    return total_sum / total_count

nested_list = [[1, 2], [3, 4], [5, 6]]
result = average_nested(nested_list)

print("Nested List:", nested_list)
print("Average of All Elements:", result)
```

```python
import numpy as np

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# 1. Prove A · A^-1 = I
A_inv = np.linalg.inv(A)
identity = np.dot(A, A_inv)

# 2. Prove AB ≠ BA
AB = np.dot(A, B)
BA = np.dot(B, A)
are_equal = np.array_equal(AB, BA)

# 3. Prove (AB)^T = B^T A^T
AB_transpose = (np.dot(A, B)).T
B_transpose_A_transpose = np.dot(B.T, A.T)
transpose_equal = np.array_equal(AB_transpose, B_transpose_A_transpose)

coeff_matrix = np.array([[2, -3, 1],
                         [1, -1, 2],
                         [3, 1, -1]])
const_vector = np.array([-1, -3, 9])

solution = np.dot(np.linalg.inv(coeff_matrix), const_vector)

print("A · A^-1:\n", identity)
print("AB:\n", AB)
print("BA:\n", BA)
print("AB equals BA?:", are_equal)
print("(AB)^T:\n", AB_transpose)
print("B^T A^T:\n", B_transpose_A_transpose)
print("(AB)^T equals B^T A^T?:", transpose_equal)
print("Solution to the system of equations [x, y, z]:", solution)
```
